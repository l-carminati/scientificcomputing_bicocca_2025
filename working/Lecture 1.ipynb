## Q11: Rock-Paper-Scissors

Implement a set of games of rock-paper-scissors against the computer.  

  * Ask for input from the user ("rock", "paper", or "scissors") and the randomly select one of these for the computer's play.
  * Announce who won.
  * Keep playing until a player says that they no longer want to play.
  * When all games are done, print out how many games were won by the player and by the computer 

```python
import random

moves = {
    "rock": 1, "paper": 2, "scissors": 3,
    "r": 1, "p": 2, "s": 3
}


def play_the_game(pc_move, user_move):
    outcome = "tie"

    if pc_move == user_move:
        outcome = "tie"
    elif pc_move > user_move or pc_move - user_move == -2:
        outcome = "pc"
    else:
        outcome = "user"

    return outcome

    
#MAIN
pc_wins = 0
user_wins = 0
tot_game = 0

willing_to_play = True

while willing_to_play:

    tot_game += 1
    pc_move = random.randrange(1, 4)  # 1,2,3

    print("What will you play? (write rock paper scissors or r p s)")
    user_input = input()

    if user_input not in moves:
        print("Invalid move, please try again\n")
        continue

    user_move = moves[user_input]

    outcome = play_the_game(pc_move, user_move)

    if outcome == "pc":
        print("Computer wins this round")
        pc_wins += 1
    elif outcome == "user":
        print("You win this round")
        user_wins += 1
    else:
        print("It's a tie!")

    print("Do you still wanna play? (y/n)")
    if input() == "n":
        willing_to_play = False

print("\nFinal Results:")
print(f"You won {user_wins} times")
print(f"Computer won {pc_wins} times")
print(f"Tied {tot_game - user_wins - pc_wins} times on {tot_game} games played")
```

## Q13: Calendar events

We want to keep a schedule of events.  We will do this by creating a class called `Day`.  It is sketched out below.  A `Day` holds a list of events and has methods that allow you to add an delete events.  Our events will be instances of a class `Event`, which holds the time, location, and description of the event.

Finally, we can keep track of a list of all the `Day`s for which we have events to make our schedule.

Fill in these classes and write some code to demonstrate their use:

```python
import random

class Day:
    """a single day keeping track of the events scheduled"""
    def __init__(self, month, day, year):
        # store the month, day, and year as data in the class
        self.month = month   
        self.day = day
        self.year = year
        # keep track of the events
        self.events = []

    def __repr__(self):
        return f"Day({self.day}/{self.month}/{self.year}, {len(self.events)} events)"
        
    def add_event(self, name, time=None, location=None):
        new_event = Event(name, time, location)
        self.events.append(new_event)
        #pass
    
    def delete_event(self, name):
        self.events = [evt for evt in self.events if evt.name != name]
        #pass
    def compute_work_day_length(self):
        self.events.sort(key=lambda evt: evt.time)
        work_day_length = self.events[len(self.events)-1].time - self.events[0].time + self.events[len(self.events)-1].duration
        print(f"Event0 at {self.events[0].time} EventN at {self.events[len(self.events)-1].time} for a day work length of {work_day_length}")
    
    
class Event:
    """a single event in our calendar"""
    def __init__(self, name, time=9, location=None, duration=1):
        self.name = name
        self.time = time
        self.location = location
        self.duration = duration


#MAIN
day = 17
month = 11
year = 2025
N_days = 7

calendar = []

for i in range(day, day + N_days): 
    calendar.append(Day(month, i, year))

# add lunch every day at 12
for day in calendar:
    day.add_event("Lunch", time=12, location="U6")


random_ev = ["Lecture", "Lab", "Meeting"]
random_loc = ["U1", "U2", "U3", "U4"]

# add random events
for day in calendar:
    time = 8 + random.randrange(0,8)
    while time < 19:
        
        if time == 12:
            continue
        else:
            random_ev_index = random.randrange(0, len(random_ev))
            random_loc_index = random.randrange(0, len(random_loc))
            day.add_event(random_ev[random_ev_index], time, random_loc[random_loc_index])
        time += random.randrange(0,8)

for day in calendar:
    day.compute_work_day_length()
```
